```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("kor_gas.csv")
```

```python
plt.figure(figsize=(30, 10))
plt.plot(df.loc[df.type == 'A', 'supply'].reset_index().supply)

plt.yscale('log')
```

```python
import matplotlib.pyplot as plt

for t in df.type[:1]:
    tmp = df.loc[df.type == t]
    for h in range(1, 25):
        tmp_supply = tmp.loc[tmp.time == h, 'supply']
        plt.plot(tmp_supply.reset_index().supply)
        plt.yscale('log')
        plt.show()
```
